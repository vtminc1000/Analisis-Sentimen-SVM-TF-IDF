```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset_sentimen_nuklir_final.csv")

df.head()
```

```python
sentimen_count = df["predicted_label"].value_counts()

sentimen_count
```

```python
plt.figure(figsize=(6,4))
sentimen_count.plot(kind="bar")
plt.title("Distribusi Sentimen terhadap Nuklir")
plt.xlabel("Sentimen")
plt.ylabel("Jumlah Komentar")
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sentimen_count.plot(kind="pie", autopct="%1.1f%%")
plt.title("Persentase Sentimen terhadap Nuklir")
plt.ylabel("")
plt.show()
```
