```python
import pandas as pd
import re
import string
```

```python
df = pd.read_csv("komentar_youtube_nuklir.csv")

df.head()
```

```python
df.columns
```

```python
print("Jumlah data awal:", len(df))
```

```python
df = df[["text"]].copy()

df.head()
```

```python
def clean_text(text):
    text = str(text)

    # lowercase
    text = text.lower()

    # hapus URL
    text = re.sub(r"http\S+|www\S+|https\S+", " ", text)

    # hapus mention
    text = re.sub(r"@\w+", " ", text)

    # hapus hashtag tapi tetap simpan katanya
    text = re.sub(r"#", "", text)

    # hapus emoji dan karakter non-ASCII
    text = text.encode("ascii", "ignore").decode("ascii")

    # hapus angka
    text = re.sub(r"\d+", " ", text)

    # hapus tanda baca
    text = text.translate(str.maketrans("", "", string.punctuation))

    # hapus spasi berlebih
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
df["clean_text"] = df["text"].apply(clean_text)

df[["text", "clean_text"]].head(10)
```

```python
df = df[df["clean_text"] != ""]

print("Jumlah setelah hapus kosong:", len(df))
```

```python
df = df.drop_duplicates(subset="clean_text")

print("Jumlah setelah hapus duplikat:", len(df))
```

```python
df[["text", "clean_text"]].sample(10, random_state=42)
```

```python
df.to_csv("komentar_youtube_nuklir_clean.csv", index=False, encoding="utf-8-sig")
```
