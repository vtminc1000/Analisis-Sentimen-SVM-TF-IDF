```python
import pandas as pd
```

```python
df = pd.read_csv("komentar_youtube_nuklir_clean.csv")

df.head()
```

```python
sample_label = df.sample(n=300, random_state=42)

sample_label = sample_label[["text", "clean_text"]].copy()
sample_label["label"] = ""

sample_label.head()
```

```python
sample_label.to_csv("sample_labeling_300.csv", index=False, encoding="utf-8-sig")
```
